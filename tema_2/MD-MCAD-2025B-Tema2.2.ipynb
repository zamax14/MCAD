## 2. Selección de variables para el modelo de regresión logística

```python
bank_data_vars = bank_data.columns.values.tolist()
Y = ['y']
X = [v for v in bank_data_vars if v not in Y]
```

```python
# Delimitar el número de variables a utilizar para el modelo de regresión logistica.
n = 10
```

```python
from sklearn import datasets
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
rfe = RFE(lr, n_features_to_select=n)
rfe = rfe.fit(bank_data[X], bank_data[Y].values.ravel())
```

```python
print(rfe.support_)
```

```python
print(rfe.ranking_)
```

```python
z= list(zip(bank_data_vars,rfe.support_, rfe.ranking_))
z
```

```python
valcols = np.append(rfe.support_,False)
cols = bank_data.iloc[:,valcols].columns.values.tolist()
```

```python
cols
```

```python
X = bank_data[cols]
Y = bank_data['y']
```

## 3. Implementación del modelo en Python con statsmodel.api

```python
import statsmodels.api as sm
```

```python
logit_model = sm.Logit(Y, X)
result = logit_model.fit()
```

```python
result.summary2()
```

## 4. Implementación del modelo en Python con scikit-learn

```python
from sklearn import linear_model
```

```python
logit_model = linear_model.LogisticRegression()
logit_model.fit(X,Y)
```

```python
logit_model.intercept_
```

```python
logit_model.coef_
```

```python
logit_model.score(X,Y)
```

```python
pd.DataFrame({"Variables": X.columns, "Coeficientes": list(np.transpose(logit_model.coef_))})
```
