Tenemos que el valor de "Pi" se puede aproximar mediante "$4*P$". Para nuestro caso, la probabilidad "P" la calcularemos mediante una simulacion en la que se arrojen puntos aleatorios dentro de nuestro cuadrado y observar quienes caen dentro del circulo. Para esto consideremos lo siguiente:

Podemos calcular la distancia que hay entre dos puntos usando la formula "d".

- Para nuestro punto inicial $(x_1,y_1)$ utilizaremos el centro de la figura localizado en $(L/2,L/2)$.
- Mientras que el punto final $(x_2,y_2)$ sera un punto aleatorio que generaremos mediante un sistema aleatorio con distribucion uniforme.

Si la distancia de nuestro punto con respecto al centro es menor o igual al radio, quiere decir que estamos dentro del circulo, por lo que manejaremos un contador para almacenar cada punto que cumpla con esto.

Por lo tanto nuestra formula queda como:
$$pi = 4 * \frac{puntos\_dentro}{numero\_muestras}$$

## Implementación

```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Semilla del experimento (Opcional)
np.random.seed(10)
```

```python
def montecarlo_pi(L, muestras):
    r = L/2
    centro = [L/2, L/2]

    x = np.random.uniform(0, L, muestras)
    y = np.random.uniform(0, L, muestras)

    dentro = 0
    
    for i in range(muestras):
        if np.sqrt((x[i]-centro[0])**2 + (y[i]-centro[1])**2) <= r:
            dentro += 1
    prob_circulo = dentro/muestras
    pi = 4*prob_circulo
    return pi
```
