# A. Muestreo Aleatorio

```python
import numpy as np
from IPython.display import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

* La estadística tiene como proposito la estimación de ciertos **parámetros** característicos de una ***población*** en específico.
* Algunos de estos parámetros poblacionales incluyen:
    * **Media poblacional ($\mu$)**
    * **Desviación típica poblacional ($\sigma$)**
    * **Proporción poblacional ($p$)**

* En la práctica, suele ser *virtualmente imposible* analizar a todos y cada uno de los elementos que conforman una población, por lo cuál también resulta imposible cálcular directamente dichos parámetros poblacionales.
* La solución habitual empleada por los estadísticos para poder estimar estos parámetros consta en en análisis de un *subconjunto de elementos* tomados en forma *aleatoria* de la población total. A este subconjunto de elementos se le conoce como ***muestra poblacional***.


$X_i$ = {$x_1$, $x_2$, $x_3$,..., $x_n$} (siendo '$n$' el tamaño de la muestra)


* Al tomar una muestra poblacional se espera que dicha conjunto de elementos manifieste características similares a las de la población de la cuál fue tomada (distribución, media, desviación típica, etc). A estás cáracteristicas (sustraidas con respecto de la muestra poblacional) se les conoce como **estadísticos**, e incluyen:
    * **Media muestral ($\bar{X}$)**
    * **Desviación típica muestral ($\tilde{s}$)**
    * **Proporción muestral ($\hat{p}$)**

* Bajo ciertas condiciones, dichos estadísticos nos permiten *inferir* los **parámetros estadísticos** de la población total.

# B. Teorema del Límite Central
* El Teorema del Límite Central (**TLC**) o Teorema Central del Límite es un teorema fundamental de la estadística que indica que, en condiciones muy generales, si $S_n$ es la suma de $n$ variables aleatorias independientes con media '$\mu$' y varianza no nula pero finita '$\sigma$', entonces la función de distribución de $S_n$ **"se aproxima bien"** a una **distribución normal** (**distribución Gaussiana**) de media $n\mu$ y desviación típica $\sigma\sqrt{n}$, esto siempre y cuando ´$n$´ sea lo suficientemente grande; es decir:


$S_j,_n=\sum_{i=1}^n{(x_i)}\in N(n\mu,\sigma\sqrt{n})$


```python
# Generar una población 'P' a partír de algún tipo de distribución.
P = np.random.uniform(15,30,100000)
# Calcular la media "mu" y desviación típica "sigma" de "P"
mu = np.mean(P)
sigma = np.std(P)
# Mostrar Histograma de Frecuencias de 'P' e imprimir 'mu' y 'sigma'
plt.hist(P)
print("mu = ", mu, ",", "sigma = ", sigma)
```

```python
# Tomar un conjunto de "m" muestras poblacionales de tamaño "n"
S_dist = []
m = 1000
n = 10000
for i in range(m):
    Xi = np.random.choice(P,n)
    Si = np.sum(Xi)
    S_dist.append(Si)
S_dist
```

```python
# Calcular la media de la distribución 'S_dist'
Mu_S = np.mean(S_dist)
# Calcular la desviación típica de la distribución 'S_dist'
sigma_S = np.std(S_dist)
# Mostrar Histograma de Frecuencias de 'S_dist' e imprimir su media y desviación estandar
plt.hist(S_dist,bins=20)
print("Mu_S = ", Mu_S, ", ", "sigma_S = ", sigma_S)
```

```python
print("mu = ", mu, ", ", "sigma = ", sigma)
print("Mu_S = ", Mu_S, ", ", "sigma_S = ", sigma_S)
```

```python
print(n*mu)
print(sigma*np.sqrt(n))
```

```python
print(Mu_S/n)
print(sigma_S/np.sqrt(n))
```

## B.1. Estimación de la Medía Poblacional mediante el TLC
* El TLC es especialmente útil si se quiere estimar la **media poblacional**.
* Especificamente, si se tomara una muestra aleatoria de la población y se calculara su respectiva **media muestral**, entonces, según el TLC, dicho estadístico debera pertencer a una distribución aproximadamente normal denominada **distribución muestral del la media**, tal que:


$\bar{X_i}=\frac{1}{n}\sum_{i=1}^n{(x_i)}\in N(\mu_{\bar{X}},\sigma_{\bar{X}})$


* Si el tamaño de la muestra '$n$' es lo suficientemente grande, entonces el valor promedio de la distribución muestral de la media ($\mu_{\bar{X}}$) será un buen estimador del parámetro de media poblacional ($\mu$); es decir:


$\mu_{\bar{X}} = \mu$


* En forma similar, considerando un tamaño de la muestra '$n$' apropiado, la **desviación típica de la distribución muestral de la media** ($\sigma_{\bar{X}}$) deberá ser igual a la **desviación típica poblacional** ($\sigma$) dividida entre la raíz cuadrada de '$n$', es decir:


$\sigma_{\bar{X}} = \frac{\sigma}{\sqrt{n}}$

```python
# Generar una población 'P' a partír de algún tipo de distribución.
P = np.random.uniform(15,30,100000)
# Calcular la media "mu" y desviación típica "sigma" de "P"
mu = np.mean(P)
sigma = np.std(P)
# Mostrar Histograma de Frecuencias de 'P' e imprimir 'mu' y 'sigma'
plt.hist(P)
print("mu = ", mu, ",", "sigma = ", sigma)
```

```python
# Tomar un conjunto de "m" muestras poblacionales de tamaño "n"
Xmean_dist = []
m = 1000
n = 10000
for i in range(m):
    Xi = np.random.choice(P,n)
    Xmean = np.sum(Xi)/n
    Xmean_dist.append(Xmean)
Xmean_dist
```

```python
# Calcular la media de la distribución 'S_dist'
Xmean_mu = np.mean(Xmean_dist)
# Calcular la desviación típica de la distribución 'S_dist'
Xmean_sigma = np.std(Xmean_dist)
# Mostrar Histograma de Frecuencias de 'S_dist' e imprimir su media y desviación estandar
plt.hist(Xmean_dist,30)
print("mu = ", mu, ",", "sigma = ", sigma)
print("Xmean_mu = ", Xmean_mu, ", ", "Xmean_sigma = ", Xmean_sigma)
```
