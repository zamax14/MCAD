* En el área de la estadística, la **correlación** (o **dependencia**) representa la relación estadística (sea o no causal) existente entre dos variables aleatorias $x$ e $y$.
* En general, se considera que dos variables aleatorias $x$ e $y$ están correlacionadas cuando los valores de una de ellas varían sistemáticamente con respecto a los valores de la otra; si por ejemplo se encontrase que la variable $y$ manifiesta algún tipo de variación con respecto de la variable $x$, entonces dicha variación podria ser descrita como:

<div align="center">
    <font, size=5>
        $y=f(x)$
    </font>
</div>

* La relación entre dos variables aleatorias queda representada mediante la línea de mejor ajuste, trazada a partir de un conjunto de observaciones. Los principales componentes elementales de una *línea de ajuste* y, por lo tanto, de una correlación, son la *fuerza*, el *sentido* y la *forma*:
    * **Fuerza**: mide el grado en que la línea representa a las observaciones. si la nube es estrecha y alargada, se representa por una línea recta, lo que indica que la relación es fuerte; si la nube de puntos tiene una tendencia elíptica o circular, la relación es débil.
    * **Sentido**: mide la dirección de variación de los valores de una variable aleatoria "$y$" con respecto de otra variable aleatoria "$x$": si al crecer los valores de $x$ lo hacen los de $y$, la relación es directa (*pendiente positiva*); si al crecer los valores de $x$ disminuyen los de $y$, la relación es inversa (*pendiente negativa*).
    * **Forma**: establece el tipo de línea que define el mejor ajuste: la *línea recta*, la *curva monotónica* o la *curva no monotónica*.

## Coeficiente de Correlación de Pearson

* El **coeficiente de correlación de Pearson** ($r_{xy}$) es una medida que permite cuantificar la dependencia lineal que existe entre dos variables aleatorias cuantitativas $x$ e $y$. Dicha medida esta dada por:

<div align="center">
    <font size="5">
        $r_{xy}=\frac{\sum_{i=1}^n{(x_i-\bar{x})(y_i-\bar{y})}}{\sqrt{\sum_{i=1}^n{(x_i-\bar{x})^2}\sum_{i=1}^n{(y_i-\bar{y})^2}}}$
    </font>
</div>

* La correlación de Person es independiente de la escala de medida de las variables analizadas (a diferencia de la *covarianza*).

* El valor del coeficiente de correlación varía en el intervalo $[-1,1]$, indicando el signo el sentido de la relación:
    * Si $r = 1$, existe una correlación *positiva perfecta*; es decir, existe una dependencia total entre las dos variables tal que al aumentar una de ellas, la otra también lo hace en proporción constante (lo cual se denominada *relación directa*).
    * Si $0 < r < 1$, existe una correlación *positiva*.
    * Si $r = 0$, no existe relación lineal entre las variables estudiadas; sin embargo, no necesariamente implica que las variables sean independientes, pues pudieran existir relaciones *no lineales* entre estas.
    * Si $-1 < r < 0$, existe una correlación *negativa*.
    * Si $r = -1$, existe una correlación *negativa perfecta*; es decir, existe una dependencia total entre las dos variables tal que al aumentar una de ellas, la otra disminuye en proporción constante (lo cual se denomina *relación inversa*).

## Ejemplo: Correlación entre Variables Cuantitativas

```python
import pandas as pd
import numpy as np
```

```python
# Cargaremos el dataset de nombre "Advertising.csv" como un dataframe
data_ads = pd.read_csv("C:/Users/neosb/Downloads/python-ml-course-master/datasets/ads/Advertising.csv")
data_ads.head(10)
```

```python
# Longitud (número de filas) del dataframe
len(data_ads)
```

```python
import matplotlib.pyplot as plt
```

```python
# Definir una función para calcular el coeficiente de correlación respecto a dos variables
def coef_corr(df, v1, v2):
    A = (df[v1]-np.mean(df[v1]))*(df[v2]-np.mean(df[v2]))
    B = (df[v1]-np.mean(df[v1]))**2
    C = (df[v2]-np.mean(df[v2]))**2
    r_xy = sum(A)/np.sqrt(sum(B)*sum(C))
    return r_xy
```

```python
var1 = "TV"
var2 = "Sales"
r_xy = coef_corr(data_ads, var1, var2)
print("Coeficiente de Correlación de ", var1,"vs",var2, "= ", r_xy)
```

```python
# Grafica "TV" vs "Sales"
plt.plot(data_ads["TV"], data_ads["Sales"], "ro")
plt.title("Gasto en publicidad de TV vs Ventas del Producto")
```

```python
var1 = "Radio"
var2 = "Sales"
r_xy = coef_corr(data_ads, var1, var2)
print("Coeficiente de Correlación de ", var1,"vs",var2, "= ", r_xy)
```

```python
# Grafica "Radio" vs "Sales"
plt.plot(data_ads["Radio"], data_ads["Sales"], "go")
plt.title("Gasto en publicidad de Radio vs Ventas del Producto")
```

# Multicolinealidad

* La **multicolinealidad** es la relación de dependencia lineal fuerte entre más de dos variables explicativas en una regresión múltiple. En otras palabras, la multicolinealidad implica la existencia de una *correlación* alta entre más de dos variables explicativas.
* Si bien la multicolinealidad no reduce las capacidades promedio de un modelo predictivo, ciertamente puede llevar a la estimación de coeficientes que *no son estadisticamente significativos*.
* Uno de los estadísticos más útiles para determinar el grado de multicolinealidad que existe entre un conjunto de variables explicativas en un modelo de regresión lineal múltiple es el llamado **Factor de Inflación de la Varianza (VIF)**, el cuál está dado según la siguiente expresión:

<div align="center">
	<font size=5>
        $VIF = \frac{1}{1-R^2}$
	<font/>
<div/>

* donde $R^2$ corresponde al coeficiente de determinación calculado con respecto a un modelo de regresión ENTRE VARIABLES EXPLICATIVAS.

```python
import statsmodels.formula.api as smf
```

```python
# Hagamos el cálculo del "VIF" para tres casos distintos:

# Newspaper ~ TV + Radio
lm_news = smf.ols(formula="Newspaper~TV+Radio", data=data_ads).fit()
r2_news = lm_news.rsquared
VIF_news = 1/(1-r2_news)
print("VIF_Newspaper = ", VIF_news)

# TV ~ Newspaper + Radio
lm_tv = smf.ols(formula="TV~Newspaper+Radio", data=data_ads).fit()
r2_tv = lm_tv.rsquared
VIF_tv = 1/(1-r2_tv)
print("VIF_TV = ", VIF_tv)

# Radio ~ TV + Newspaper
lm_radio = smf.ols(formula="Radio~TV+Newspaper", data=data_ads).fit()
r2_radio = lm_radio.rsquared
VIF_radio = 1/(1-r2_radio)
print("VIF_Radio = ", VIF_radio)

# Sales~TV+Radio+Newspaper
lm = smf.ols(formula="Sales~TV+Radio+Newspaper", data=data_ads).fit()
r2 = lm.rsquared
VIF_Sales = 1/(1-r2)
print("VIF_Sales = ", VIF_Sales)
```

* Al calcular el valor $VIF$, es importante observar lo siguiente:
    * Si $VIF=1$: Las variables no están correlacionadas.
    * Si $VIF<5$: Las variables tienen una correlación moderada.
    * Si $VIF>5$: Las variables están altamente correlacionadas.

### NOTAS:
* Si una variable independiente tiene un grado de correlación importante con respecto de las demás, pudiera ser conveniente descartarla del modelo de predicción.
* Para el caso anterior, lo mas pertienente es descartar la variable "Newspaper" del modelo, pues es la que tiene el mayor indice de multicolinealidad.
